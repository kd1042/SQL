```python
# Generate synthetic data of a retail store, using faker,pandas for practicing analyzing the data and finding patterns in sql and power BI
```

```python
import pandas as pd
import faker as f
import random as ran
```

```python
import datetime
```

```python

fake = f.Faker()
```

```python
# create customer data
customers = []
for i in range(1000) :
    customer = {'id' : i + 1,
                'name' : fake.name(),
                'phone' : fake.phone_number(),
                'email' : fake.email()
                }
    customers.append(customer)
```

```python
customers
```

```python
cust_table = pd.DataFrame(customers,columns=['id','name','phone','email'] )
```

```python
cust_table
```

```python
import faker_commerce
```

```python
fake.add_provider(faker_commerce.Provider)
```

```python
#create product data
products = []
for i in range(0,200) :
    product = {'poduct_id' : int(i + 1),
        'product_name' : fake.ecommerce_name(),
        'sale_price' : ran.randrange(200,500),
        'purchase_price' : ran.randrange(100,400),
        'supplier_id':ran.randrange(1,20),}
    products.append(product)
```

```python
products
```

```python
prod_table = pd.DataFrame(products,columns = ['product_id','product_name','sale_price','purchase_price','supplier_id'])
prod_table
```

```python
#order_table
orders = []
for i in range(5000) :
    order = {'order_id' : i + 1,
            'product_id' : ran.randrange(0,199),
            'quantity' : ran.randrange(1,10),
            'customer_id' : ran.randrange(1,1000),
            'date':str(fake.date_between_dates(date_start =datetime.date(2021,1,1),date_end =datetime.date(2023,12,31)))}
    orders.append(order)
orders
```

```python
order_table = pd.DataFrame(orders, columns = ['order_id','product_id','quantity','customer_id','date'])
order_table
```

```python
#suppliers
suppliers = []
for i in range(20) :
    supplier = {'supplier_id' : i + 1,
                'name' : fake.company(),
                'address' : fake.address(),
                'phone' : fake.phone_number(),
                'email' : fake.email(),
               }
    suppliers.append(supplier)
suppliers
    
```

```python
supplier_table = pd.DataFrame(suppliers,columns = ['supplier_id','name','address','phone','email'])
```

```python
supplier_table
```
